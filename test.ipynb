```python
import pandas as pd
import json
import ast
from pathlib import Path

def check_zero_embeddings(csv_file):
    """Check for all-zero embeddings in a CSV file"""
    print(f"\nAnalyzing: {csv_file}")
    print("=" * 80)
    
    df = pd.read_csv(csv_file)
    zero_count = 0
    zero_rows = []
    
    for idx, row in df.iterrows():
        embedding_str = row['embedding']
        
        # Parse the embedding string to a list
        try:
            embedding = ast.literal_eval(embedding_str)
        except:
            print(f"Error parsing embedding at row {idx}")
            continue
        
        # Check if all values are 0 or very close to 0
        if all(abs(val) < 1e-10 for val in embedding):
            zero_count += 1
            zero_rows.append({
                'row_index': idx,
                'table': row['table'],
                'col': row['col'],
                'header': row['header']
            })
    
    print(f"Total rows: {len(df)}")
    print(f"All-zero embeddings found: {zero_count}")
    
    if zero_rows:
        print("\nZero embeddings found at:")
        for item in zero_rows:
            print(f"  Row {item['row_index']}: {item['table']}, Col {item['col']}, Header: {item['header']}")
    else:
        print("\n✓ No all-zero embeddings detected")
    
    return zero_count, zero_rows


# Define paths
base_path = Path("/home/kpanag/vscode/cta_full_pipeline/SOTABV2forSemTab2023/CTA-SCH-R1/gt/supplementary_files")

snowflake_files = [
    base_path / "train_snowflake-arctic-embed2:568m_AVG:True.csv",
    base_path / "test_snowflake-arctic-embed2:568m_AVG:True.csv",
    base_path / "val_snowflake-arctic-embed2:568m_AVG:True.csv"
]

# snowflake_files = [
#     base_path / "train_nomic-embed-text-v2-moe:latest_AVG:False.csv",
#     base_path / "test_nomic-embed-text-v2-moe:latest_AVG:False.csv",
#     base_path / "val_nomic-embed-text-v2-moe:latest_AVG:False.csv"
# ]

# Check each file
all_results = {}
for csv_file in snowflake_files:
    if csv_file.exists():
        zero_count, zero_rows = check_zero_embeddings(str(csv_file))
        all_results[csv_file.name] = {'count': zero_count, 'rows': zero_rows}
    else:
        print(f"\nFile not found: {csv_file}")

# Summary
print("\n" + "=" * 80)
print("SUMMARY")
print("=" * 80)
for filename, result in all_results.items():
    print(f"{filename}: {result['count']} all-zero embeddings")
```

```python
import pandas as pd
from collections import defaultdict, Counter

def get_initial_list(df1):
    nameList = []
    for i in range(len(df1)):
        name = df1['table'][i].split('_')[0]
        nameList.append(name)
    nameList = list(set(nameList))    
    return nameList

def get_domains_with_counts(GT_PATH1, GT_PATH2):
    df1 = pd.read_csv(GT_PATH1)
    df2 = pd.read_csv(GT_PATH2)

    domain_counts = defaultdict(Counter)

    for df in [df1, df2]:
        for i in range(len(df)):
            domain = df['table'][i].split('_')[0]
            attr = df['header'][i]
            domain_counts[domain][attr] += 1

    return domain_counts

def get_domains(GT_PATH1,GT_PATH2):
    df1 = pd.read_csv(GT_PATH1)
    df2 = pd.read_csv(GT_PATH2)

    names = get_initial_list(df1)
    dict = {name: [] for name in names}

    for i in range(len(df1)):
        dict[df1['table'][i].split('_')[0]].append(df1['header'][i])

    for i in range(len(df2)):
        dict[df2['table'][i].split('_')[0]].append(df2['header'][i])

    for key in dict:
        dict[key] = list(set(dict[key]))
    
    #print(dict)
    return dict
#dom_dict = get_domains("SOTABV2forSemTab2023/CTA-SCH-R2/gt/sotab_cta_train_round2.csv","SOTABV2forSemTab2023/CTA-SCH-R2/gt/sotab_cta_validation_round2.csv")
dom_dict = get_domains_with_counts("SOTABV2forSemTab2023/CTA-SCH-R2/gt/sotab_cta_train_round2.csv","SOTABV2forSemTab2023/CTA-SCH-R2/gt/sotab_cta_validation_round2.csv")
print(dom_dict)
```
